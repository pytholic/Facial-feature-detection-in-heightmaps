# Data loading and preprocessing

```python
# Load image
img_test = cv2.imread("./data_heightmap/100.png")
```

```python
plt.figure(figsize = (10,10))
plt.imshow(img_test)
```

```python
dim = (224, 224)
resized = cv2.resize(img_test, dim, interpolation = cv2.INTER_AREA)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(resized)
```

Detection failed in some cased. Setting the **illumination** fixed the issue. So we change the illumination via opencv as follows.

```python
# Setting illumination for better detection
img_exp = resized.copy()
```

```python
def adjust_gamma(image, gamma=1.0):

    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")

    return cv2.LUT(image, table)
```

```python
gamma = 0.4
```

```python
adjusted = adjust_gamma(img_exp, gamma=gamma)
```

```python
fig = plt.figure(figsize = (10,10))
ax1 = plt.subplot(1,2,1)
ax1.title.set_text("Original image")
plt.imshow(img_exp)
ax2 = plt.subplot(1,2,2)
ax2.title.set_text("Adjusted image")
plt.imshow(adjusted)
```

```python
enhanced = cv2.detailEnhance(adjusted, sigma_s=10, sigma_r=0.15)
```

```python
fig = plt.figure(figsize = (15,15))
ax1 = plt.subplot(1,3,1)
ax1.title.set_text("Original image")
plt.imshow(img_exp)
ax2 = plt.subplot(1,3,2)
ax2.title.set_text("Adjusted image")
plt.imshow(adjusted)
ax3 = plt.subplot(1,3,3)
ax3.title.set_text("Enhanced image")
plt.imshow(enhanced)
```

# Apply model

```python
### LANDMARK DETECTION ###
```

```python
img = enhanced.copy()
#img = resized.copy()
#img = edges.copy()
```

```python
mp_face_mesh = mediapipe.solutions.face_mesh
```

```python
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, refine_landmarks=True)
```

```python
results = face_mesh.process(img)
```

```python
print(len(results.multi_face_landmarks))
```

```python
landmarks = results.multi_face_landmarks[0]
```

```python
# print(landmarks.landmark[0])
print(len(landmarks.landmark))
```

```python
#xs = []; ys = []; zs = []

for landmark in landmarks.landmark:
    x = landmark.x
    y = landmark.y
    #z = landmark.z
    
#     xs.append(x)
#     ys.append(y)
#     zs.append(z)
    
    relative_x = int(x * img.shape[1])
    relative_y = int(y * img.shape[0]) # To denormalize values
    
    cv2.circle(img, (relative_x, relative_y), radius=2, color=(0, 0, 255), thickness=-1)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(img[:,:,::-1]) # Reversing to change it to RGB from BGR
```

```python
### CONTOUR DETECTION ###
```

```python
img = resized.copy()

for source_idx, target_idx in mp_face_mesh.FACEMESH_CONTOURS:
    
    source = landmarks.landmark[source_idx]
    target = landmarks.landmark[target_idx]
    
    relative_source = (int(source.x * img.shape[1]), int(source.y * img.shape[0]))
    relative_target = (int(target.x * img.shape[1]), int(target.y * img.shape[0]))
    
    cv2.line(img, relative_source, relative_target, color=(0, 0, 255), thickness=2)
    
plt.figure(figsize = (10,10))
plt.imshow(img[:,:,::-1]) # Reversing to change it to RGB from BGR
```

```python
### FACE MESH DETECTION ###
```

```python
img = resized.copy()

for source_idx, target_idx in mp_face_mesh.FACEMESH_TESSELATION:
    
    source = landmarks.landmark[source_idx]
    target = landmarks.landmark[target_idx]
    
    relative_source = (int(source.x * img.shape[1]), int(source.y * img.shape[0]))
    relative_target = (int(target.x * img.shape[1]), int(target.y * img.shape[0]))
    
    cv2.line(img, relative_source, relative_target, color=(255, 255, 255), thickness=1)
    
plt.figure(figsize = (10,10))
plt.imshow(img[:,:,::-1]) # Reversing to change it to RGB from BGR
```
